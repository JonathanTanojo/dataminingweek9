# Association Rule Mining

Association Rule Mining adalah sebuah metode akan pembelajaran mesin berbasis akan aturan untuk menemukan akan hubungan menarik antara variabel dalam database besar.

Asscociation Rule adalah pernyataan "if-then" Statements, yang dimana akan membantu menunjukan kemungkinan hubungan antara item data, dalam kumpulan data besar dalam berbagai jenis akan database, yang dimana mining memiliki beberapa aturan yang dimana membantu untuk menemukan korelasi penjualan dalam data transaksional atau dalam kumpulan data - data medis
